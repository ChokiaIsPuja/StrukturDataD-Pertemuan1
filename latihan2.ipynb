1. List Kosong

```python
list = []
```

2. input Data

```python
for i in range(10):
  angka = int(input("Masukan Angka :"))
  list.append(angka)
```

3. tampilkan list, nilai tertinggi, terendah, dan rata rata

```python
print("List:", list)
print("Nilai Tertinggi:", max(list))
print("Nilai Rata-Rata:", sum(list)/len(list))
print("Nilai Terkecil", min(list))
```

4. tampilkan graph

```python
import matplotlib.pyplot as plt
plt.bar(range(len(list)), list)
plt.title("Grafik Array")
plt.xlabel("Indeks")
plt.ylabel("Nilai")
plt.show()
```
